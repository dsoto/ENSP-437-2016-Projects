# Design Strategy

- Our srategy was to keep the inside temperature between 65 and 75 degrees Farenheit. To do this we wanted an R-Value that would maintain a comfortable temperature. We added grass on top of the roof to increase the insulation where the sun would be hitting the box most. Also, since plants need a lot of sun to be able to grow, we thought that the grass roof would help absorb more of the sun's energy instead of all of it being absorbed directly by the box itself. Inside, we added about an inch of cotton between the wall of cardboard and foam. Again, we thought it would imitate the insulation of an actual house more than just using multiple pieces of foam alone. We made sure it was packed in tightly to minimize the amount of air between the walls. We also let the foam on the inside face outward so that the inner walls were white hoping it would help keep the temperature from rising too high. We only put in one south-facing window because we wanted to be able to let in enough sunlight so that the inside would be able to get warm, but we also did not want to lose all of that heat during the night either. Lastly, we let the radiant barriers on the outside foam face outwards because we thought it would help reflect some of the sun's rays rather than all of the energy being absorbed into the material.

2) To determine the amount of insulation needed we kept adjusting the R-Value until we got a UA value that would keep the amount of thermal loss to a minimum. 
We found the R-Value of the cardboard, foam, cotton, and airfilm and multiplied by the area of the box to figure out the first UA value. Then to add in the grass roof's insulation, we seperately calculated the R-Value of soil and grass and multiplied by the area of the roof to get another UA value. In order to get the total UA value of the box, we added both UA values. 

```python
from pint import UnitRegistry
u = UnitRegistry()
area = 7.33 * u.feet**2
u.define('R_US = foot**2 * delta_degF * hour / BTU')
wall_r_value = (1 + 3.5 + 3.4 + 0.17 + 0.68 + 0.425) * u.R_US
UA_product = area / wall_r_value
print('Our UA product estimate is', UA_product.to(u.BTU/u.hour/u.delta_degF))
print('Our UA product estimate is', UA_product.to(u.watt/u.delta_degC))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('supa-stars.csv', parse_dates=True, index_col=0)
data['t_diff'] = data['t_in_C'] - data['t_out_C']
data[['t_in_C', 't_out_C', 't_diff']].plot()
plt.title('Run 1')
plt.ylabel('Temperature (C)')
plt.show()
```
