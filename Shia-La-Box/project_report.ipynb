We figured that with this extra insulation, what we lost through the window would be to our benefit as conduction through two layers of insulation would be significantly reduced.

```python
from pint import UnitRegistry
u = UnitRegistry()
area = 10 * u.feet**2
u.define('R_US = foot**2 * delta_degF * hour / BTU')
wall_r_value = (0.17 + 0.5 + 0.68) * u.R_US
UA_product = area / wall_r_value
print('Our UA product estimate is', UA_product.to(u.BTU/u.hour/u.delta_degF))
print('Our UA product estimate is', UA_product.to(u.watt/u.delta_degC))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('shia_la_box.csv', parse_dates=True, index_col=0)
data['t_diff'] = data['t_in_C'] - data['t_out_C']
data[['t_in_C', 't_out_C', 't_diff']].plot()
plt.title('Shia_LaBox')
plt.ylabel('Temperature (C)')
plt.show() 
```
